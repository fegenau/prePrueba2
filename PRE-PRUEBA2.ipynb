## Intentaremos realizar una consolidacion de la data para que al momento de analizarla sea mas facil

```python
# Consolidamos los datos de las peliculas de Star Wars
data = pd.concat([data1, data2, data3, data4, data5, data6, data7, data8, data9])
```

```python
# Visualizamos los datos consolidados
print(data)
```

Al notar que existe tres copias de datos de las peliculas, intentaremos eliminar los datos duplicados y solamente dejar una fila por columna, para evitar problemas en el análisis de los datos.

```python
# Buscaremos datos duplicados en los nombres de las peliculas
duplicates = data.duplicated(subset=['Title'])
print(duplicates)
```

```python
# Eliminamos los datos duplicados segun el titulo, dejando solamente una fila por pelicula
data = data.drop_duplicates(subset=['Title'])
```

```python
# Visualizamos los datos limpios
print(data)
```

## Ahora que hemos limpiado los datos, podemos proceder a realizar un análisis exploratorio de los mismos.

```python
# Exploración de datos utilizando visualizaciones básicas con matplotlib o seaborn
data.head(10)
```

```python
# Visualizaremos un heat map para ver un mapa de calor de las peliculas que mas ganaro y menos ganaron
#Para ello primero tenemos que cambiar el tipo de dato de la columna BoxOffice a float
data['BoxOffice'] = data['BoxOffice'].str.replace('$', '')
data['BoxOffice'] = data['BoxOffice'].str.replace(',', '.')
data['BoxOffice'] = data['BoxOffice'].astype(float)
```

```python
# Visualizamos el tipo de dato de la columna BoxOffice
print(data['BoxOffice'].dtype)
```

```python
# Transformamos la columna BoxOffice a int
data['BoxOffice'] = data['BoxOffice'].astype(int)
```

```python
# Visualizamos el mapa de calor de las peliculas que mas ganaron y menos ganaron
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Mapa de Calor de las Peliculas de Star Wars')
plt.show()
```
